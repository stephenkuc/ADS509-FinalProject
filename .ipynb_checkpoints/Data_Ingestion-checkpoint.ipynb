Overall, we want to find tweets mentioning Elon Musk and/or Twitter by handle, and analyze the sentiment of users' tweets that mention them, and later, predict for that sentiment.

Elon's handle is @elonmusk

Twitter's handle is @twitter

We will likely use the API.search_tweets() function to find these.

```python
queryTopic = "@elonmusk OR @twitter lang:en"

tweets = []

num_tweets = 25000

# Replace the limit=1000 with the maximum number of Tweets you want

for tweet in tweepy.Paginator(client.search_recent_tweets, query=queryTopic, tweet_fields=['context_annotations', 'created_at'],
                              max_results=100).flatten(limit=num_tweets):
        tweets.append(tweet.text)
```

```python
# saving these onto local machine

path = os.getcwd()

tweets_file = path + "\\tweet_data.txt"

with open(tweets_file, 'w') as f:
    f.write(json.dumps(tweets))
```
